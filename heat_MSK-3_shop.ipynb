```python
df_customer_volume = df.groupby(['Клиент','Магазин'])['CG'].sum().reset_index()
df_customer_volume
```

```python
quantiles = [0, 0.25, 0.5, 0.75, 1]

# calculate the quantile values for the purchase volume
quantile_values = df_customer_volume['CG'].quantile(quantiles)

# create a new column for classification
df_customer_volume['classification_GMV'] = pd.cut(df_customer_volume['CG'], 
                                             bins=quantile_values, 
                                             labels=['Low', 'Medium', 'High', 'Very High'])
```
